```python
df.drop(['Unnamed: 0','New_Price'], axis=1,inplace=True)
df
```

```python
df['Mileage'] = df['Mileage'].str.replace('km/kg','')
df['Mileage'] = df['Mileage'].str.replace('kmpl','')
df['Engine'] = df['Engine'].str.replace('CC','')
df['Power'] = df['Power'].str.replace('bhp','')
df
```

```python
df['Mileage'] = df['Mileage'].astype(float)
df['Engine'] = df['Engine'].astype(float)
df['Power'] = df['Power'].str.replace("null",'0')   # null cannot be converted to float. so..
df['Power'] = df['Power'].astype(float)
```

```python
df.isna().sum()
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
df['Engine'] = df['Engine'].fillna(df['Engine'].mode()[0])
df['Power'] = df['Power'].fillna(df['Power'].mean())
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])
df.isna().sum()
```

```python
df['brand'] = df['Name'].apply(lambda x: x.split()[0])
df
```

```python
# Label encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = []
df['Owner_Type'] = le.fit_transform(df['Owner_Type'])
df['Transmission'] = le.fit_transform(df['Transmission'])
df['Fuel_Type'] = le.fit_transform(df['Fuel_Type'])
df['Location'] = le.fit_transform(df['Location'])
df['brand'] = le.fit_transform(df['brand'])

df.drop(['Name'], axis=1, inplace=True)

df
```

```python
# le.inverse_transform(df['Owner_Type']) gives orginal data.

#  When using a LabelEncoder (or any other transformation object) from scikit-learn to transform data,
#  if you transform multiple columns using the same object, it can lead to issues when trying to inverse-transform the data.
# it's recommended to use separate instances of the LabelEncoder (or similar objects) for each column,
#  so each column has its own independent encoding and can be correctly inverse-transformed.
```

```python
X = df.iloc[:, :-2]
X['brand'] = df['brand']   # or just drop the price column and add rest to X
X
```

```python
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("r2_score",r2_score(y_test,y_pred))
print("mean_absolute_error ", mean_absolute_error(y_test,y_pred) )
print("mean_squared_error ", mean_squared_error(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kr = KNeighborsRegressor()
kr.fit(X_train,y_train)
y_pred = kr.predict(X_test)
print("r2_score",r2_score(y_test,y_pred))
print("mean_absolute_error ", mean_absolute_error(y_test,y_pred) )
print("mean_squared_error ", mean_squared_error(y_test,y_pred))
```
